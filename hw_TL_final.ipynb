```python
import transformers
from transformers import pipeline

sentiment_pipeline = pipeline("sentiment-analysis")
data = ( ["Отличный товар",
            "Мне не понравилось",
            "Прекрасно выглядит, ничего не меняйте!",
            "Ничего не понял",
            "У меня не работала розетка, в остальном все отлично"] )
sentiment_pipeline(data)
```
